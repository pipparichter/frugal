```python
def table_2(dataset_df:pd.DataFrame):

    table_df = pd.DataFrame(index=dataset_df.genome_id.unique())
    table_df['phylum'] = dataset_df.groupby('genome_id').phylum.first()
    table_df['gc_percent'] = dataset_df.groupby('genome_id').gc_percent.first()
    table_df['genome_size'] = dataset_df.groupby('genome_id').apply(len, include_groups=False)
    table_df['n_matches'] = dataset_df.groupby('genome_id').match.sum()
    table_df['n_hypothetical_matches'] = dataset_df.groupby('genome_id').apply(lambda df : ((df.top_hit_product == 'hypothetical protein') & df.match).sum(), include_groups=False)
    table_df['n_exact_matches'] = dataset_df.groupby('genome_id').exact_match.sum()
    table_df['n_extended'] = dataset_df.groupby('genome_id').extended.sum()
    table_df['n_truncated'] = dataset_df.groupby('genome_id').truncated.sum()
    table_df['n_pseudogenes'] = dataset_df.groupby('genome_id').pseudogene.sum()
    table_df['n_conflicts'] = dataset_df.groupby('genome_id').conflict.sum()
    table_df['n_intergenic'] = dataset_df.groupby('genome_id').intergenic.sum()

    totals = table_df.drop(columns=['gc_percent', 'phylum']).sum(axis=0)
    for col in [col for col in totals.index if col.startswith('n_')]:
        print(col.replace('n_', '', count=1), get_percent(totals.loc[col], totals.genome_size), f'(n={totals.loc[col]})')

    for col in [col for col in table_df.columns if col.startswith('n_')]:
        table_df[col.replace('n_', '', count=1)] = [f'{n} ({get_percent(n, total)})' for n, total in zip(table_df[col], table_df.genome_size)]

    table_df = table_df.sort_values('gc_percent')
    table_df.index = table_df.index.map(dataset_df.groupby('genome_id').species.first())

    save_table(table_df.drop(columns=[col for col in table_df.columns if col.startswith('n_')]))


table_2(dataset_df)
```

```python
# Want to at least ballpark the number of spurious Prodigal predictions. Base this on (1) unlikely conflict with annotated sequences in the reference, (2) conflict with non-coding genes, 
# (3) reported false positive values on better-studied genomes. 


def figure(results_df:pd.DataFrame):

    fig, (ax_top, ax_bottom) = get_split_figure((0, 0.1), (0.78, 0.82))

    dimonaco_df = pd.read_csv('../data/dimonaco_2020.csv') # Load in data from "No one tool to rule them all" paper.
    dimonaco_df['false_positive_rate'] = (dimonaco_df.num_cds - dimonaco_df.num_match) / (dimonaco_df.num_cds)
    dimonaco_df = dimonaco_df.sort_values('false_positive_rate', ascending=False)
    palette = {'Prodigal':'gray', 'GeneMarkS-2':'lightgray'}
    sns.scatterplot(dimonaco_df, x='gc_content', y='false_positive_rate', hue='tool', legend=True, palette=palette, edgecolor='black', lw=1, ax=ax_top)
    sns.scatterplot(dimonaco_df, x='gc_content', y='false_positive_rate', hue='tool', palette=palette, edgecolor='black', lw=1, ax=ax_bottom, legend=False)


    # Don't include the high-error outlier when fitting the line. 
    m, b = np.polyfit(dimonaco_df[dimonaco_df.tool == 'Prodigal']['gc_content'].values[1:], dimonaco_df[dimonaco_df.tool == 'Prodigal']['false_positive_rate'].values[1:], deg=1)
    x = np.linspace(30, 70, 100)
    y = m * x + b 
    sns.lineplot(x=x, y=y, ls='--', color='black', alpha=0.5, ax=ax_top)
    sns.lineplot(x=x, y=y, ls='--', color='black', alpha=0.5, ax=ax_bottom)

    figure_df = pd.DataFrame(index=pd.Index(results_df.species.unique()))
    figure_df['gc_content'] = results_df.groupby('species').gc_percent.first()
    figure_df['n_spurious'] = results_df.groupby('species').spurious.sum()
    figure_df['total'] = results_df.groupby('species').size()
    figure_df['false_positive_rate'] = figure_df.n_spurious / figure_df['total']

    # figure_df['num_n_terminal_extension'] = dataset_df[is_annotated_n_terminal_extension(dataset_df)].groupby('species').apply(len, include_groups=False)
    # figure_df['total'] = dataset_df[is_annotated_match(dataset_df)].groupby('species').apply(len, include_groups=False)
    # figure_df['n_terminal_extension_rate'] = dataset_df.groupby('species').apply(lambda df : is_annotated_n_terminal_extension(df).sum() / is_annotated_match(df).sum(), include_groups=False)
    x = figure_df.gc_content.values
    sns.scatterplot(figure_df, x='gc_content', y='false_positive_rate', color='indianred', legend=True, s=10)
    ax_top.set_ylabel('')

    false_postive_rate = dimonaco_df.false_positive_rate.min() # Get the minimum false positive rate reported by Dimonaco et. al. 
    n_false_positive =  int(false_postive_rate * len(dataset_df))

    print('figure: Num. false positives lower bound based on Dimonico et. al data:', n_false_positive, f'({get_percent(n_false_positive, len(dataset_df))})')

    save_figure(fig)

figure(results_df)

```

```python
def figure(dataset_df:pd.DataFrame):

    total = len(dataset_df)

    # Filter for the conflicts. 
    figure_df = dataset_df[dataset_df.conflict & (dataset_df.top_hit_product != 'hypothetical protein')].copy()
    print('figure: Num. conflicts with non-hypothetical proteins:', len(figure_df))

    is_antiparallel_cds_conflict = lambda df : (~df.same_strand) & (df.conflict) & (df.top_hit_feature == 'CDS') # Include nested. 
    is_non_coding_conflict = lambda df : (df.conflict) & (df.top_hit_feature != 'CDS')
    is_tandem_cds_conflict = lambda df : (df.same_strand) & (df.conflict) & (df.top_hit_feature == 'CDS') # Include nested. 

    conditions = [is_antiparallel_cds_conflict(figure_df), is_tandem_cds_conflict(figure_df), is_non_coding_conflict(figure_df)]
    assert sum([mask.sum() for mask in conditions]) == len(figure_df), f'figure: Sum of masks should equal the total number of conflicts, but {sum([mask.sum() for mask in conditions])} != {len(figure_df)}.'
    categories = ['antiparallel', 'tandem', 'noncoding']
    figure_df['category'] = np.select(conditions, categories, default='none')

    fig, ax = get_figure()
    sns.ecdfplot(figure_df, hue='category', palette='Grays', x='overlap_length')

    ax.axvline(x=60, ls='--', color='black')
    # text = get_text('CDS > 60', (figure_df[is_cds_conflict(figure_df)].overlap_length > 60).sum(), len(figure_df))
    text = get_text('tandem > 60', (figure_df[figure_df.category == 'tandem'].overlap_length > 60).sum(), len(figure_df))
    text += get_text('antiparallel > 60', (figure_df[figure_df.category == 'antiparallel'].overlap_length > 60).sum(), len(figure_df))
    text += get_text('noncoding', (figure_df.category == 'noncoding').sum(), len(figure_df))
    ax.text(0.5, 0.4, text, transform=ax.transAxes)
    ax.set_xlim(xmax=500, xmin=0)
    ax.set_ylabel('proportion')

    n_false_positive = (figure_df[figure_df.category == 'antiparallel'].overlap_length > 60).sum() + (figure_df.category == 'noncoding').sum()
    print('figure: Num. false positive lower bound:', n_false_positive, f'({get_percent(n_false_positive, total)})')

    save_figure(fig)

figure(dataset_df)
```

```python
def figure_2(results_df:pd.DataFrame, path:str='../figures/dataset-2_figure-2.pdf'):

    fig, (ax_top, ax_bottom) = get_split_figure((0, 600), (48700, 48850))

    n_real_hypothetical_match = (is_top_hit_hypothetical(results_df) & results_df.real & results_df.match).sum()
    n_aligned_hypothetical_match = (is_top_hit_hypothetical(results_df) & results_df.aligned & results_df.match).sum()
    n_real_aligned_hypothetical_match = (is_top_hit_hypothetical(results_df) & results_df.aligned & results_df.match & results_df.real).sum()
    n_hypothetical_match = (is_top_hit_hypothetical(results_df) & results_df.match).sum()
    print('figure: Num. real hypothetical matches:', n_real_hypothetical_match, f'({get_percent(n_real_hypothetical_match, n_hypothetical_match)})')
    print('figure: Num. hypothetical matches with a SwissProt homolog:', n_aligned_hypothetical_match, f'({get_percent(n_aligned_hypothetical_match, n_hypothetical_match)})')
    print('figure: Num. real hypothetical matches with a SwissProt homolog:', n_real_aligned_hypothetical_match, f'({get_percent(n_real_aligned_hypothetical_match, n_aligned_hypothetical_match)})')

    figure_df = results_df.copy()
    figure_df['category'] = np.select([results_df.match, results_df.pseudogene, results_df.intergenic, results_df.conflict], ['match', 'pseudogene', 'intergenic', 'conflict'], default='none')
    categories = ['match', 'conflict', 'intergenic', 'pseudogene']
    figure_df['category'] = pd.Categorical(figure_df.category, categories, ordered=True)

    width = 0.25
    offsets = {'spurious':-0.3, 'real':0, 'uncertain':0.3}
    palette = {'spurious':'indianred', 'real':'darkseagreen', 'uncertain':'lightgray'}

    for model_label, df in figure_df.groupby('model_label'):
        ax_df = pd.DataFrame(index=categories)
        ax_df['total'] = df.category.value_counts()
        ax_df['fraction'] = df.groupby('category', observed=True).apply(lambda df : df.top_hit_unsupported.sum() / len(df), include_groups=False)
        ax_df['unsupported'] = ax_df['total'] * ax_df['fraction']
        ax_df['confirmed'] = ax_df['total'] * (1 - ax_df['fraction'])

        positions = np.arange(df.category.nunique())
        ax_top.bar(positions + offsets[model_label], ax_df.unsupported, color=palette[model_label], edgecolor='black', hatch='///', width=width)
        ax_top.bar(positions + offsets[model_label], ax_df.confirmed, bottom=ax_df.unsupported, color=palette[model_label], edgecolor='black', width=width, label=model_label)
        ax_bottom.bar(positions + offsets[model_label], ax_df.unsupported, color=palette[model_label], edgecolor='black', hatch='///', width=width)
        ax_bottom.bar(positions + offsets[model_label], ax_df.confirmed, bottom=ax_df.unsupported, color=palette[model_label], edgecolor='black', width=width, label=model_label)

        for x, total in zip(positions + offsets[model_label], ax_df.total):
            if total < 600:
                ax_bottom.text(x, total, str(total), ha='center', fontsize='small', va='bottom')
        ax_bottom.set_xticks(np.arange(4), ax_df.index)

    ax_top.legend()

    save_figure(fig, path=path)

figure_2(results_df)
```
